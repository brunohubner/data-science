```python
# Importação das bibliotecas
import pandas as pd
import seaborn as srn
from os import path
```

```python
# Carregamento da base de dados
base = pd.read_csv(path.join("..", "data", "dados.csv"), sep=';')
base.shape
```

```python
base.head()
```

```python
srn.distplot(base.PIB, bins=10, axlabel="PIB").set_title("PIB")
```

```python
srn.distplot(base.VALOREMPENHO, bins=10,
             axlabel="Valor do Empenho").set_title("Empenho")
```

```python
pib = base.sort_values("PIB").head(10)
pib = pib.iloc[:, 1:3]
pib.plot.bar(x="MUNICIPIO", y="PIB")
```

```python
valor_empenho = base.sort_values("VALOREMPENHO").head(10)
valor_empenho = valor_empenho.iloc[:, [1, 3]]
valor_empenho.plot.bar(x="MUNICIPIO", y="VALOREMPENHO")
```
